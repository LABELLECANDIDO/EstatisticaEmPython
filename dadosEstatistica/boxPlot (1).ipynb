```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/drive')
dados_metaverso = pd.read_csv('/content/drive/MyDrive/metaverse_transactions_dataset.csv')

# Transações de Baixo Risco
risco_baixo_norte_america = dados_metaverso[(dados_metaverso['anomaly'] == 'low_risk') & (dados_metaverso['location_region'] == 'North America')]

# Transações de Alto Risco
risco_alto_norte_america = dados_metaverso[(dados_metaverso['anomaly'] == 'high_risk') & (dados_metaverso['location_region'] == 'North America')]

# Transações de Risco Moderado
risco_moderado_norte_america = dados_metaverso[(dados_metaverso['anomaly'] == 'moderate_risk') & (dados_metaverso['location_region'] == 'North America')]

# Crie um gráfico com os boxplots
sns.boxplot(
    x = "anomaly",
    y = "amount",
    showmeans=True,  # Mostra a média de cada nível de risco
    data=pd.concat([risco_baixo_norte_america, risco_alto_norte_america, risco_moderado_norte_america])
)

# Rotacionar etiquetas do eixo X
plt.xticks(rotation=45)

# Mostrar o gráfico
plt.show()
```
